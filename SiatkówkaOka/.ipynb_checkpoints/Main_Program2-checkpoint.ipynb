### Machine learning - 4.0

```python
import numpy as np
import cv2


scale_percent = 1000
width = int(image.prepared.shape[1] * scale_percent / 100)
height = int(image.prepared.shape[0] * scale_percent / 100)
dim = (width, height)

Tree = Training.Classifier(cv2.resize(np.array(image.prepared, dtype='uint8'), dim, interpolation=cv2.INTER_AREA),
                           cv2.resize(np.array(image.holdOut, dtype='uint8'), dim, interpolation=cv2.INTER_AREA))


```

```python
image2 = IMG.Image()
image2.imageProcess(paths[3], hoPaths[3])
Tree.testing(cv2.resize(np.array(image2.prepared, dtype='uint8'), dim, interpolation=cv2.INTER_AREA),
             cv2.resize(np.array(image2.holdOut, dtype='uint8'), dim, interpolation=cv2.INTER_AREA))

```

```python

Effect.checkEffectiveness(Tree.predicted.reshape(1920, 1998), Tree.target_test.reshape(1920, 1998))
Tree.show3(cv2.resize(np.array(image2.prepared, dtype='uint8'), (1920, 1998), interpolation=cv2.INTER_AREA))
```
