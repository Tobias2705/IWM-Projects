```python
#Importy
import sys
import numpy as np
import cv2
from ZdjecieNaSinogram import zdjNAsinogram
from matplotlib import pyplot as plt
from skimage import data, io
```

```python
#Zdefiniowane klasy
class Zdjecie:    
    def __init__(self, zdj):
        self.oryginal = zdj
    
    def pokazZdjecie(self, plot):
        fig, plots = plt.subplots(1,2)
        plots[0].imshow(self.oryginal, cmap='gray')
        plots[1].imshow(plot, cmap='gray')
        plt.show()
    
    def zdjecieNaSinogram(self):
        self.sinogram, self.linie = zdjNAsinogram(self.oryginal, krok=4, liczba_detektorow=90, rozwartosc=90)
        print("Zdjęcie oryginalne vs zdjęcie sinogramu")
        self.pokazZdjecie(self.sinogram)
        
        return self.sinogram
```

```python
#Pozostałe funkcje
def tomograf(zdjecie):
    zdj = Zdjecie(zdjecie)
    zdj.zdjecieNaSinogram()
```

```python
#Main
zdjecie = np.zeros([200,200])
#Dla zdj 01
#zdjecie[13:187, 25:174] = cv2.cvtColor(io.imread("Zdjecia/Po_konwersji/01.png"), cv2.COLOR_RGBA2GRAY)

zdjecie[25:175, 25:175] = cv2.cvtColor(io.imread("Zdjecia/Po_konwersji/02.png"), cv2.COLOR_RGBA2GRAY)

tomograf(zdjecie)
```
